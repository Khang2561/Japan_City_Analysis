```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.available
plt.style.use('seaborn-whitegrid')
```

```python
df1 = pd.read_csv("japan_gdp.csv")  # read file 1
df1.head(1000)
```

```python
df_japan = df1[(df1['Entity']=='Japan') & (df1['Year']>=1960) ]
df_america = df1[(df1['Entity']=='United States') & (df1['Year']>=1960) ]
df_Russia = df1[(df1['Entity']=='Russia') & (df1['Year']>=1960) ]
df_Russia
```

```python
plt.plot(df_japan["Year"], df_japan["GDP"]/100000000, color="blue", label="GDP_Japan");
plt.plot(df_america["Year"], df_america["GDP"]/100000000, color="red", label="GDP_USA");
plt.plot(df_Russia["Year"], df_Russia["GDP"]/100000000, color="green", label="GDP_Russia");

plt.axis("tight")
plt.title ("GPD Japan")

plt.legend();


plt.xlabel("Year");
plt.ylabel("GDP (x 100 million)");
```
