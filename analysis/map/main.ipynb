```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("data_population.csv")#read file 1
df1.head()
```

```python
df2 = pd.read_csv("country-cities-data.csv")#read file 1
df2.head()
```

```python
cities = pd.merge(df1,df2,on = 'City')#connect file 1 with file 2
cities.head()
```

```python
lat, lon = cities['latitude'],cities['longitude']#extract latitude and longitude
```

```python
population, area = cities['Population'],cities['Area']#extract population and area
city = cities['City']
# Remove commas from area values
area = [float(a.replace(',', '')) for a in area]

# Convert the area list to a numpy array
area = np.array(area, dtype=float)

population = [float(a.replace(',', '')) for a in population]

# Convert the area list to a numpy array
population = np.array(population, dtype=float)
```

```python
plt.style.use('seaborn')#select style
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,10));

#plot using pylot API
plt.scatter(lon,lat,c=population/100000, cmap='autumn_r',s=area/3,linewidths= 0.5,alpha = 0.5);
plt.axis('equal');
plt.xlabel('longlatitude');
plt.ylabel('latitude');
plt.colorbar(label='population');
plt.title("japan of city");

```
